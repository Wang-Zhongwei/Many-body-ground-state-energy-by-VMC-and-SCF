```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
dat = np.loadtxt('../data/hf/annealing/Be_annealing.dat', unpack=True)

# first column 
alpha = dat[0]
beta = dat[1]
energy = dat[2]
```

```python
# plot energy as a function of alpha and as a function of beta subplot

# plot the trajectory of alpha and beta and energy vs iteration number as subplot
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,5))
# improve definition
# plot with arrow on line 
ax1.plot(alpha, beta, 'o-', markersize=1.5)
ax1.set_xlabel(r'$\alpha$')
ax1.set_ylabel(r'$\beta$')
ax1.set_title(r'Parameter trajectory')
# set dot size of ax2
ax2.plot(energy, 'ro-', markersize=1.5)
ax2.set_xlabel(r'iter')
ax2.set_ylabel(r'$E$')
ax2.set_title(r'Energy vs iteration')

plt.show()
```

```python
fig.savefig('../data/hf/annealing/Be_annealing1.png', dpi=300)
```
